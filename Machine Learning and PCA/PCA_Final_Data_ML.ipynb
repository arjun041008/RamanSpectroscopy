```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px
import plotly.graph_objects as go
from sklearn.preprocessing import Normalizer
from sklearn.decomposition import PCA
```

```python
df = pd.read_excel("/content/drive/MyDrive/RSI_Google Colabs/final_augmented_data.xlsx")
df.index = df["Sample_ID"]
df.drop("Sample_ID", axis=1, inplace=True)
df.head(5)
```

```python
df_pca = df[df.columns[1:]]

df_pca = pd.DataFrame(Normalizer().fit_transform(df_pca), columns = df_pca.columns, index = df_pca.index)

df_pca.head(5)
```

```python
pca = PCA(n_components = 3)

df_pca = pd.DataFrame(pca.fit_transform(df_pca), columns = ["PC1", "PC2", "PC3"], index = df_pca.index)

df_pca.head(5)
```

```python
df_pca["Presence of Bacteria"] = df["Presence of Bacteria"]
df_pca.head(5)
```

```python
df_pca.to_excel("/content/drive/MyDrive/RSI_Google Colabs/PCA_Data.xlsx")
```

```python
fig = px.scatter_3d(df_pca, x='PC1', y='PC2', z='PC3',
                    color='Presence of Bacteria',
                    title='PCA Analysis of Milk Type')

fig.show()
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.naive_bayes import GaussianNB
from sklearn.gaussian_process import GaussianProcessClassifier
from sklearn.gaussian_process.kernels import RBF
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report, roc_curve, auc, precision_score, recall_score, f1_score, roc_auc_score
from sklearn.model_selection import StratifiedKFold, cross_val_score
from sklearn.svm import SVC
from sklearn.neighbors import KNeighborsClassifier
```

```python
kf = StratifiedKFold(n_splits=5)

y = df_pca["Presence of Bacteria"]
x = df_pca.drop("Presence of Bacteria", axis=1)
```

```python
log_reg = LogisticRegression()

scores = cross_val_score(log_reg, x, y, cv=kf)
print("Logistic Regression Average Accuracy" , scores.mean()*100 , "%")
```

```python
gauss = GaussianProcessClassifier(kernel = 1.0 * RBF(1.0))

scores = cross_val_score(gauss, x, y, cv=kf)
print("Gaussian Process Average Accuracy" , scores.mean()*100 , "%")
```

```python
bayes = GaussianNB()

scores = cross_val_score(bayes, x, y, cv=kf)
print("Naive Bayes Average Accuracy" , scores.mean()*100 , "%")
```

```python
rf = RandomForestClassifier()

scores = cross_val_score(rf, x, y, cv=kf)
print("Random Forest Average Accuracy" , scores.mean()*100 , "%")
```

```python
svm = SVC()

scores = cross_val_score(svm, x, y, cv=kf)
print("Support Vector Machine Average Accuracy" , scores.mean()*100 , "%")
```

```python
knn = KNeighborsClassifier()

scores = cross_val_score(knn, x, y, cv=kf)
print("K-Nearest Neighbors Average Accuracy" , scores.mean()*100 , "%")
```

```python
model = GaussianNB()


accuracies, precisions, recalls, f1s, aucs = [], [], [], [], []


all_y_true = []
all_y_pred = []
all_y_scores = []

for train_idx, val_idx in kf.split(x, y):
    X_train, X_val = x.iloc[train_idx], x.iloc[val_idx]
    y_train, y_val = y.iloc[train_idx], y.iloc[val_idx]

    model.fit(X_train, y_train)
    y_pred = model.predict(X_val)

    if hasattr(model, "predict_proba"):
        y_score = model.predict_proba(X_val)[:, 1]
    else:
        y_score = model.decision_function(X_val)


    all_y_true.extend(y_val)
    all_y_pred.extend(y_pred)
    all_y_scores.extend(y_score)


    accuracies.append(accuracy_score(y_val, y_pred))
    precisions.append(precision_score(y_val, y_pred, zero_division=0))
    recalls.append(recall_score(y_val, y_pred, zero_division=0))
    f1s.append(f1_score(y_val, y_pred, zero_division=0))
    aucs.append(roc_auc_score(y_val, y_score))

print("Classification Report for" , model)
print("Mean Accuracy:", np.mean(accuracies))
print("Mean Precision:", np.mean(precisions))
print("Mean Recall:", np.mean(recalls))
print("Mean F1-score:", np.mean(f1s))
print("Mean AUC:", np.mean(aucs))
```
